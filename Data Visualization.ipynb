```python
import pandas as pd
df = pd.read_csv(r"Mall_Customers.csv")
```

```python
df.head(), df.describe(), df.info()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

# line chart

```python
df[['Annual Income (k$)', 'Spending Score (1-100)']].plot.line()
```

# bar chart

```python
gen_inc_exp = round(df.groupby("Gender")[["Spending Score (1-100)", "Annual Income (k$)"]].mean(),2)
print(gen_inc_exp)
gen_inc_exp.plot(kind = 'bar')
```

# barh chart

```python
gen_inc_exp = round(df.groupby(["Gender"])[["Spending Score (1-100)", "Annual Income (k$)"]].mean(),2)
print(gen_inc_exp)
gen_inc_exp.plot(kind = 'barh', alpha = 0.6, stacked = True)
```

# histogram chart

```python
df[["Age","Spending Score (1-100)", "Annual Income (k$)"]].plot.hist(bins = 50, alpha = 0.5)
```

# box plot

```python
df[["Age", "Spending Score (1-100)", "Annual Income (k$)"]].plot.box()
```

# Kernel density - KDE

```python
df[["Gender", "Age", "Spending Score (1-100)", "Annual Income (k$)"]].plot.kde()
```

# Area plot

```python
df[["Age", "Spending Score (1-100)", "Annual Income (k$)"]].plot.area(alpha=0.5)
```

# Pie Plot

```python
gen_inc_exp.plot(kind = 'pie', subplots = True)
```

# Scatter Plot

```python
df.plot(kind = 'scatter', y = "Age", x = "Annual Income (k$)")
print("Correlation",df['Age'].corr(df["Annual Income (k$)"]))
```

```python
df.plot(kind = 'scatter', y = "Spending Score (1-100)", x = "Annual Income (k$)")
print("Correlation",df["Spending Score (1-100)"].corr(df["Annual Income (k$)"]))
```

```python
df.plot(kind = 'scatter', y = "Age", x = "Spending Score (1-100)")
print("Correlation",df["Age"].corr(df["Spending Score (1-100)"]))
```

# Hexbin Plot

```python
df.plot.hexbin(x = 'Age', y = "Spending Score (1-100)", gridsize=15, figsize = (7,5))
df.plot(kind = 'hexbin', y = "Age", x = "Annual Income (k$)", gridsize=15, figsize = (7,5))
df.plot.hexbin(x = "Annual Income (k$)", y = "Spending Score (1-100)", gridsize=15, figsize = (7,5))
```
